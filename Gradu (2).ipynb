```python
import pandas as pd
import matplotlib.pyplot as plt
import mpltern
```

```python
data = pd.read_csv("Batch_1&2_formula_comp_6Si.csv", skiprows = 1, nrows = 75)
rap9 = data.iloc[0:15]
rap7peg = data.iloc[15:25]
rap7host = data.iloc[25:35]
lan1peg = data.iloc[35:45]
lan1host = data.iloc[45:55]
lan6peg = data.iloc[55:65]
lan6host = data.iloc[65:75]

data2 = pd.read_csv("February_all_formula_comp_6Si.csv", skiprows = 1, nrows = 102)


lev4 = data2.iloc[0:15]
hei3 = data2.iloc[15:21]
hei2a = data2.iloc[21:29]
kau1 = data2.iloc[29:39]
kau2 = data2.iloc[39:54]
mat2wall = data2.iloc[54:63]
mat2peg = data2.iloc[63:72]
mat1 = data2.iloc[72:82]
jan1wall = data2.iloc[82:92]
jan1peg = data2.iloc[92:102] 
```

```python
# Primary tourmaline groups, X-site
ax = plt.subplot(projection="ternary")
# Data here
ax.scatter(rap9["Ca(X)"], rap9["X-vacancy"], 
           (rap9["Na(X)"] + rap9["K(X)"]), 
           label = "Rapasaaret (core)", marker = '*', c = 'tab:orange', s = 20)
ax.scatter(rap7peg["Ca(X)"], rap7peg["X-vacancy"], 
           (rap7peg["Na(X)"] + rap7peg["K(X)"]), 
           label = "Rapasaaret (contact, pegmatite)", marker = '.', c = 'tab:orange')
ax.scatter(rap7host["Ca(X)"], rap7host["X-vacancy"], 
           (rap7host["Na(X)"] + rap7host["K(X)"]), 
           label = "Rapasaaret (contact, wall rock)", marker = '^', c = 'tab:orange', s = 15)
ax.scatter(lan1peg["Ca(X)"], lan1peg["X-vacancy"], 
           (lan1peg["Na(X)"] + lan1peg["K(X)"]), 
           label = "Länttä (contact, pegmatite)", marker = '.', c = 'tab:green')
ax.scatter(lan1host["Ca(X)"], lan1host["X-vacancy"], 
           (lan1host["Na(X)"] + lan1host["K(X)"]), 
           label = "Länttä (contact, wall rock)", marker = '^', c = 'tab:green', s = 15)
ax.scatter(lan6peg["Ca(X)"], lan6peg["X-vacancy"], 
           (lan6peg["Na(X)"] + lan6peg["K(X)"]), marker = '.', c = 'tab:green')
ax.scatter(lan6host["Ca(X)"], lan6host["X-vacancy"], 
           (lan6host["Na(X)"] + lan6host["K(X)"]), marker = '^', c = 'tab:green', s = 15)
ax.scatter(lev4["Ca(X)"], lev4["X-vacancy"], 
           (lev4["Na(X)"] + lev4["K(X)"]), 
           label = "Leviäkangas (contact, wall rock)", marker = '^', c = 'tab:red', s = 15)
ax.scatter(hei3["Ca(X)"], hei3["X-vacancy"], 
           (hei3["Na(X)"] + hei3["K(X)"]), 
           label = "Heikinkangas (contact, wall rock)", marker = '.', c = 'tab:purple')
ax.scatter(hei2a["Ca(X)"], hei2a["X-vacancy"], 
           (hei2a["Na(X)"] + hei2a["K(X)"]), 
           label = "Heikinkangas (core)", marker = '*', c = 'tab:purple', s = 20)
ax.scatter(jan1peg["Ca(X)"], jan1peg["X-vacancy"], 
           (jan1peg["Na(X)"] + jan1peg["K(X)"]), 
           label = "Jäneslampi (contact, pegmatite)", marker = '.', c = 'tab:pink')
ax.scatter(jan1wall["Ca(X)"], jan1wall["X-vacancy"], 
           (jan1wall["Na(X)"] + jan1wall["K(X)"]), 
           label = "Jäneslampi (contact, wall rock)", marker = '^', c = 'tab:pink', s = 15)
ax.scatter(mat2peg["Ca(X)"], mat2peg["X-vacancy"], 
           (mat2peg["Na(X)"] + mat2peg["K(X)"]), 
           label = "Matoneva (contact, pegmatite)", marker = '.', c = 'tab:blue')
ax.scatter(mat2wall["Ca(X)"], mat2wall["X-vacancy"], 
           (mat2wall["Na(X)"] + mat2wall["K(X)"]), 
           label = "Matoneva (contact, wall rock)", marker = '^', c = 'tab:blue',s = 15)
ax.scatter(mat1["Ca(X)"], mat1["X-vacancy"], 
           (mat1["Na(X)"] + mat1["K(X)"]), 
           label = "Matoneva (core)", marker = '*', c = 'tab:blue', s = 20)
ax.scatter(kau1["Ca(X)"], kau1["X-vacancy"], 
           (kau1["Na(X)"] + kau1["K(X)"]), 
           label = "Kaustinen pegmatite granite (KAU-1)", marker = '1', c = 'k')
ax.scatter(kau2["Ca(X)"], kau2["X-vacancy"], 
           (kau2["Na(X)"] + kau2["K(X)"]), label = "Kaustinen pegmatite granite (KAU-2)", marker = '2', c = 'k')


# Lines, ticks, labels etc.
ax.plot([0.5, 0.5], [0.5, 0.5], [0.0, 0.5], color ="black", linewidth = 0.7)
ax.plot([0.0, 0.5], [0.5, 0.5], [0.5, 0.5], color ="black", linewidth = 0.7)
ax.plot([0.5, 0.5], [0.0, 0.5], [0.5, 0.5], color ="black", linewidth = 0.7)
ax.taxis.set_ticks([0.25, 0.50, 0.75, 1.0])
ax.laxis.set_ticks([0.25, 0.50, 0.75, 1.0])
ax.raxis.set_ticks([0.25, 0.50, 0.75, 1.0])
ax.tick_params(labelrotation = "horizontal")
ax.set_tlabel('Ca')
ax.set_llabel('Vacancy')
ax.set_rlabel('Na (+K)')
ax.laxis.set_label_rotation_mode("horizontal")
ax.raxis.set_label_rotation_mode("horizontal")
ax.legend(bbox_to_anchor = (1.75, 1.0))
ax.text(0.55, 0.33, 0.12, 'Calcic group')
ax.text(0.15, 0.75, 0.10, 'X-vacant group')
ax.text(0.15, 0.35, 0.50, 'Alkali group')
plt.title("Tourmaline X-site classification")
plt.savefig('Tourmaline_X-site.png', dpi = 500, bbox_inches = 'tight')
```
